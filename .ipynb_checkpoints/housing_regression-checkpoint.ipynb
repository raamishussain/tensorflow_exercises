# Housing Prices Regression Exercise 

This notebook implements linear regression and dense neural network (DNN) regression with Tensorflow to predict housing prices using California housing data from the 1990 census. 

The dataset contains information about housing block groups. This exercise
will estimate the median house value for each housing block.

The dataset can be found here: http://www.liaad.up.pt/~ltorgo/Regression/DataSets.html.

```python
import numpy      as np
import pandas     as pd
import tensorflow as tf
import seaborn    as sns
import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow       import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

np.set_printoptions(precision=3, suppress=True)
```

```python
housing_data = pd.read_csv('cal_housing_clean.csv')
housing_data.head()
```

```python
# Perform 70/30 train test split and separate features and labels

train_data = housing_data.sample(frac=0.7,random_state=1)
test_data  = housing_data.drop(train_data.index)

train_feats = train_data.drop('medianHouseValue',axis=1)
train_labels = train_data['medianHouseValue']

test_feats = test_data.drop('medianHouseValue',axis=1)
test_labels = test_data['medianHouseValue']
```

```python
# Make pair plots of data using Seaborn
sns.pairplot(train_data,diag_kind='hist')
```

## Create Linear Regression Model

```python
# plotting function to plot losses
def plot_loss(history):
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch',fontsize=13)
    plt.ylabel('Error',fontsize=13)
    plt.yscale('log')
    plt.ylim(1e4,1e6)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Use normalizer to independently normalize each feature
normalizer = preprocessing.Normalization(axis=-1)
norm_train_feats = normalizer.adapt(np.array(train_feats))
```

```python
# Create Linear Model
linear_model = tf.keras.Sequential([normalizer,layers.Dense(units=6,activation='tanh'),
                                    layers.Dense(units=6,activation='tanh'),
                                    layers.Dense(units=6,activation='tanh')])
```

```python
# Configure the model by choosing optimizer and loss function
linear_model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.1),
                     loss='mean_squared_error')
```

```python
# Train model (Use 20% of training data for validation)
history = linear_model.fit(train_feats,train_labels,epochs=1000,verbose=0,validation_split = 0.2)
```

```python
# Plot losses using our plotting function
plot_loss(history)
```
